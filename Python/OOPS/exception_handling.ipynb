#### Exception
An exception in Python is an event that disrupts the normal flow of a program's execution. When an error occurs during the execution of a program, Python raises an exception, which can then be handled by the programmer. Exceptions provide a way to deal with errors and unexpected situations gracefully, rather than causing the program to crash.

##### Inorder to continue the smooth flow of execution without any interruption. we have to handle these errors

```python
# we can handle these errors using the Exception Handling

#  Basic Calculator

num1 = 10

try:
    # keeping the error possibility block of code in try block. if any errors. 
    # those will get caughted by except block without interrupting the flow of execution
    
    num2 = int(input("enter the number :"))
    print("enter the number :", num2)


    add = num1 + num2
    print("addition is : ", add)

    div = num1 / num2
    print("division is : ", div)

except Exception as e:  # Exception class handles the all types of runtime errors
    print(e)

mul = num1 * num2
print("multiplication is : ", mul)
```

##### NameError: num3 is not defined. which can be handled by general exception

```python
# we can handle these errors using the Exception Handling

#  Basic Calculator

num1 = 10

try:
    # keeping the error possibility block of code in try block. if any errors. 
    # those will get caughted by except block without interrupting the flow of execution
    
    # entered number abc 
    num2 = int(input("enter the number :")) 
    print("enter the number :", num2)

    add = num1 + num3           # num3 is a variable which is not defined. gives the NameError
    print("addition is : ", add)

    div = num1 / num2
    print("division is : ", div)

# Multiple except blocks
except ZeroDivisionError as e:  # handling only zerodivisionerror
    print(e)

except ValueError as e:  # handling only ValueError
    print(e)

except TypeError: # handling TypeError
    print("please enter the correct value")     # we can pass the custom messages of our own

except Exception as e:  # this handles the all exception which ever exception is not being handled will be caught here
    print(e)

mul = num1 * num2
print("multiplication is : ", mul)
```
