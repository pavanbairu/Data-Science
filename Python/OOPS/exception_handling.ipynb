#### Exception
An exception in Python is an event that disrupts the normal flow of a program's execution. When an error occurs during the execution of a program, Python raises an exception, which can then be handled by the programmer. Exceptions provide a way to deal with errors and unexpected situations gracefully, rather than causing the program to crash.

##### finally block in exceptions -> finally block execute always irrespective the exception
##### used to perform cleanup actions, such as closing files or releasing resources, ensuring they are executed even if an exception occurs in the try block.

```python
# we can handle these errors using the Exception Handling

#  Basic Calculator

num1 = 10

try:
    # keeping the error possibility block of code in try block. if any errors. 
    # those will get caughted by except block without interrupting the flow of execution
    
    # entered number abc 
    num2 = int(input("enter the number :")) 
    print("enter the number :", num2)

    add = num1 + num2           # num3 is a variable which is not defined. gives the NameError
    print("addition is : ", add)

    div = num1 / num2
    print("division is : ", div)

# Multiple except blocks
except Exception as e:  # keeping the main exception at first
    print(e)

except ZeroDivisionError as e:  # handling only zerodivisionerror
    print("please enter the number other than 0")

except ValueError as e:  # handling only ValueError
    print("please enter the correct value")

except TypeError: # handling TypeError
    print("please enter the correct type of value")     # we can pass the custom messages of our own

else:
    print("coming into the else block")

finally:
    print("in finally block")

print("outside")
```

Note: finally block is executed with exception

##### DB connection example

```python
class DB:
    def conn(self):
        print ("connected")
    
    def close(self):
        print ("connection closed")
    def get(self):
        print ("here are your 500 records")
```

```python
db = DB()
```

```python
try:
    db.conn()       # db gets conncetd
    print("Charging 1000000 per hour")
    1/0     # raising exception here 

except ValueError as e:  
    print(e)

db.close()      # clsoing db connection out of exception block
```

Note: due to not handling the error the db connection has not got closed

```python
try:
    db.conn()       # db gets conncetd
    print("Charging 1000000 per hour")
    1/0     # raising exception here 

except ValueError as e:  
    print(e)

finally:
    db.close() #  closing db connection in side finally block
```

Note: even after raising the exception the finally block is executed and closed the db connection

```python
def Final():
    db = DB()
    try :
        db.conn()
        db.get()
        return 1
    except ValueError :
        return 0
    finally:
        db.close()
```

```python
class SalaryOutOfRange(Exception):      # inhering the main exception class

    def __init__(self, msg):
        self.msg = msg
```

##### custom exception on invalid Age, mail

```python
# creating an class for the person 
class person:

    def __init__(self, age, mail):

        self.age = age 
        self.mail = mail

    def get_age(self):
        return self.age
    
    def get_email(self):
        return self.mail
```

```python
person1 = person(-29, 'pavan')      # the age and mail are not valid we have throw an custom exception invalid age and mail
print(person1.get_age())
print(person1.get_email())
```

##### creating an exception class for age and email

```python
# Invalid Age exception class
class InvalidAgeError(Exception):

    def __init__(self, msg):
        self.msg = msg

# Invalid email exception class 
class InvalidEmailError(Exception):

    def __init__(self, msg):
        self.msg = msg
```

```python
age = person1.get_age()
age
```

```python
person1.get_email()
```

```python
try:
    
    age = person1.get_age()
    mail = person1.get_email()

    if age <= 0:    # age raises exception
        raise InvalidAgeError("Please enter the Age greater than 0")
    
    if "@" not in mail and ".com" not in mail:  # email address raise an exception if has not raised above
        raise InvalidEmailError("Please enter the valid email address")

except InvalidAgeError as e:    # raises Invalid age exception
    print(e)

except InvalidEmailError as e:  # raises Invalid mail exception
    print(e)
```

##### Handle multiple exception 

##### creating an exception class

```python
# age exception
class InvalidAgeError(Exception):

    def __init__(self, age):
        self.msg = age
        super().__init__(f"Invalid age: {age}. Age must be a positive integer.")

# Email exception    
class InvalidEmailError(Exception):

    def __init__(self, email):
        self.msg = email
        super().__init__(f"Invalid email format: {email}. Email must be in the format 'username@example.com'.")

# Email & age exception
class InvalidAgeEmailError(Exception):

    def __init__(self, age, email):
        self.email = email
        self.email = email
        super().__init__(f"Invalid email format: {email}. Email must be in the format 'username@example.com'." + "\n" + f"Invalid age: {age}. Age must be a positive integer.")
```

```python
# creating an class for the person 
class person:

    def __init__(self, age, mail):


        # raise exception here itself if email and age are not valid
        if age <= 0 and "@" not in mail and ".com" not in mail:
            raise InvalidAgeEmailError(age, mail)
        
        self.age = age
        self.mail = mail
        
        if age <= 0:    # age raises exception
            raise InvalidAgeError(age)
        
        if "@" not in mail and ".com" not in mail:  # email address raise an exception if has not raised above
            raise InvalidEmailError(mail)
```
