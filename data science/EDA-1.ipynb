```python
# Check for missing values in the dataset
# The .isnull() method returns a DataFrame of the same shape as 'dataset',
# where each element is a boolean indicating whether that element is missing (True) or not (False).
dataset.isnull()
```

```python
# Calculate the total number of missing values in each column
# The .isnull() method returns a DataFrame of the same shape as 'dataset',
# where each element is a boolean indicating whether that element is missing (True) or not (False).
# The .sum() method then sums these boolean values for each column, giving the count of missing values per column.
missing_values_per_column = dataset.isnull().sum()

# Display the count of missing values for each column
missing_values_per_column
```

```python
# Identify features (columns) with missing values
# Use a list comprehension to iterate over each column in the dataset.
# For each column, check if the sum of missing values (using .isnull().sum()) is not equal to 0.
# If a column has missing values, add it to the list 'features_with_na'.

features_with_na = [feature for feature in dataset.columns if dataset[feature].isnull().sum() != 0]

# Display the list of features with missing values
features_with_na
```

```python
# Calculate the percentage of missing values for the 'LotFrontage' feature
# Divide by the total number of entries in the 'LotFrontage' column (using len(dataset['LotFrontage'])) to get the percentage of missing values.
missing_percentage = dataset['LotFrontage'].isnull().sum() / len(dataset['LotFrontage'])

# Display the percentage of missing values
missing_percentage
```

```python
# Iterate over each feature (column) with missing values
for missing_feature in features_with_na:
    # Calculate the percentage of missing values for the current feature

    # Multiply by 100 to convert the fraction to a percentage
    # np.round(..., 5) rounds the percentage to 5 decimal places
    
    missing_percentage = np.round(dataset[missing_feature].isnull().sum() / len(dataset[missing_feature]) * 100, 5)
    
    # Print the feature name and the percentage of missing values
    print(f"{missing_feature}: {missing_percentage}% missing values")
```

```python
# Create a boxplot for the 'SalePrice' column
sns.boxplot(dataset['SalePrice'])
```

```python
data = dataset.copy()

# Iterate over the list of features that contain missing values
for index, feature in enumerate(features_with_na):
    # If the feature contains null values, replace them with 1, otherwise 0
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # Create a bar plot on the corresponding subplot
    data.groupby(feature)['SalePrice'].median().plot.bar(color='skyblue', edgecolor='black')

    plt.title(feature)
    # Adjust the layout to prevent overlap
    plt.tight_layout()

    # Show the plot
    plt.show()
```

```python
# using subplots
data = dataset.copy()

# Define the figure size and the number of subplots (rows, columns)
fig, axes = plt.subplots(nrows=10, ncols=2, figsize=(10, 20))  # Adjust rows and columns based on the number of features

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Iterate over the list of features that contain missing values
for index, feature in enumerate(features_with_na):
    # If the feature contains null values, replace them with 1, otherwise 0
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # Create a bar plot on the corresponding subplot
    data.groupby(feature)['SalePrice'].median().plot.bar(ax=axes[index], color='skyblue', edgecolor='black')
    
    # Set the title of the subplot to the feature name
    axes[index].set_title(feature)

# Remove any empty subplots
for i in range(len(features_with_na), len(axes)):
    fig.delaxes(axes[i])

# Adjust the layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

## Numerical Features

```python
dataset.info()
```

```python
# Identify numerical features in the dataset
# Use a list comprehension to iterate over each column in the dataset.
# For each column, check if its data type is not 'O' (object, typically used for strings in pandas).
# If the data type is not 'O', add the column to the list 'numerical_features'.

numerical_features = [feature for feature in dataset.columns if dataset[feature].dtype != 'O']

# Display the list of numerical features and the count of numerical features
numerical_features, len(numerical_features)
```

```python
# fetching columns for numeric features from dataset
dataset[numerical_features].head()
```

## Temporal Variables

```python
# Identify numerical features related to years or years-like information
# Use a list comprehension to filter numerical_features based on the condition:
# Check if the substring 'yr' or 'year' (case insensitive) is present in the lowercase version of each feature name.
# If the condition is met, add the feature to the list 'temp_features'.

temp_features = [feature for feature in numerical_features if 'yr' in feature.lower() or 'year' in feature.lower()]

# Display the list of numerical features related to years or years-like information
temp_features
```

```python
# Iterate over each feature related to years or years-like information
for year_feature in temp_features:
    # Print the feature name and its unique values
    print(year_feature, dataset[year_feature].unique())
```

```python
# Group the dataset by 'YrSold' and calculate the median 'SalePrice' for each year, then plot the results
dataset.groupby('YrSold')['SalePrice'].median().plot()

# Set the label for the x-axis
plt.xlabel("House sold year")

# Set the label for the y-axis
plt.ylabel("Median of House sales price")

# Set the title of the plot
plt.title("House Sold Year vs Price")

# Display the plot
plt.show()
```

```python
# Create a copy of the dataset
data = dataset.copy()

# Iterate over each time-related feature in temp_features
for time_feature in temp_features:
    
    # Skip the 'YrSold' feature to avoid calculating differences with itself
    if time_feature != 'YrSold':
        
        # Calculate the difference between 'YrSold' and the current time_feature
        data[time_feature] = data['YrSold'] - data[time_feature]

        # Create a scatter plot of the difference (or transformed) time_feature against 'SalePrice'
        plt.scatter(data[time_feature], data['SalePrice'])
        
        # Set the title of the plot
        plt.title("House Price Vs " + time_feature)
        
        # Set the label for the x-axis
        plt.xlabel(time_feature)
        
        # Set the label for the y-axis
        plt.ylabel("House Sales Price")
        
        # Display the plot
        plt.show()
```

## Discrete Features

```python
# Identify discrete numerical features in the dataset
# Use a list comprehension to iterate over each numerical feature in numerical_features.
# For each feature, check if the number of unique values is less than or equal to 25.
# If the condition is met, add the feature to the list 'discrete_features'.

discrete_features = [feature for feature in numerical_features if len(dataset[feature].unique()) <= 25]

# Display the list of discrete numerical features
discrete_features
```

```python
# Create a copy of the dataset
data = dataset.copy()

# Iterate over each discrete numerical feature in discrete_features
for discrete_feature in discrete_features:
    
    # Group the data by the current discrete feature and calculate the median 'SalePrice', then plot as a bar chart
    data.groupby(discrete_feature)['SalePrice'].median().plot.bar()
    
    # Set the label for the x-axis
    plt.xlabel(discrete_feature)
    
    # Set the label for the y-axis
    plt.ylabel('House Sales Price')
    
    # Set the title of the plot dynamically based on the discrete feature
    plt.title("House Sold Price Vs " + discrete_feature)
    
    # Display the plot
    plt.show()
```

### There is a relationship between discrete variables and Sale Price

## continous Features

```python
# Identify continuous numerical features in the dataset
# Use a list comprehension to iterate over each numerical feature in numerical_features.
# For each feature, check if it does not belong to the list of discrete_features.
# If the condition is met, add the feature to the list 'continuous_features'.

continuous_features = [feature for feature in numerical_features if feature not in discrete_features]

# Display the list of continuous numerical features
continuous_features
```

```python
# Create a copy of the dataset
data = dataset.copy()

# Iterate over each continuous numerical feature in continuous_features
for continuous_feature in continuous_features:
    
    # Group the data by the current continuous feature and calculate the median 'SalePrice', then plot as a histogram
    data.groupby(continuous_feature)['SalePrice'].median().plot.hist()
    
    # Set the label for the x-axis based on the current continuous feature
    plt.xlabel(continuous_feature)
    
    # Set the label for the y-axis as "Sales Price"
    plt.ylabel('Sales Price')
    
    # Set the title of the plot dynamically based on the current continuous feature
    plt.title('House Price vs ' + continuous_feature)
    
    # Display the plot
    plt.show()
```

```python
# Create a copy of the dataset
data = dataset.copy()

# Iterate over each continuous numerical feature in continuous_features
for feature in continuous_features:
    
    # Plot a histogram of the current feature with 25 bins
    data[feature].hist(bins=25)
    
    # Set the label for the x-axis based on the current feature
    plt.xlabel(feature)
    
    # Set the label for the y-axis as "count"
    plt.ylabel("count")
    
    # Set the title of the plot as the current feature name
    plt.title(feature)
    
    # Display the plot
    plt.show()
```

## Outliers

```python
# Create a copy of the dataset
data = dataset.copy()

# Iterate over each continuous numerical feature in continuous_features
for feature in continuous_features:
    
    # Check if the feature contains the value 0
    if 0 in data[feature].unique():
        # If the feature contains 0, skip log transformation to avoid undefined results
        pass
    else:
        # If the feature does not contain 0, apply logarithm transformation
        data[feature] = np.log(data[feature])
        
        # Plot a boxplot of the transformed feature
        data.boxplot(column=feature)
        
        # Set the label for the y-axis based on the current feature
        plt.ylabel(feature)
        
        # Set the title of the plot to indicate it shows outliers in the transformed feature
        plt.title("Outliers in " + feature)
        
        # Display the plot
        plt.show()
```

## Categorical Features

```python
# Identifying categorical features based on their data type
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtype == 'O']

# Displaying the list of categorical features
categorical_features
```

```python
# Iterate over each categorical feature
for feature in categorical_features:
    
    # Count the number of unique categories in the current feature
    num_categories = len(dataset[feature].unique())
    
    # Print the result
    print("The number of categories in", feature, "is equal to", num_categories)
```

```python
data = dataset.copy()

# Iterate over each categorical feature
for feature in categorical_features:
    
    # Group data by the current categorical feature and plot median SalePrice
    data.groupby(feature)['SalePrice'].median().plot.bar()
    
    # Adding labels and title to the plot
    plt.title("Sales Price Vs " + feature)
    plt.xlabel(feature)
    plt.ylabel("Median Sales Price")
    
    # Display the plot
    plt.show()
```
