```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.DataFrame({'Category': ['A','A','B', 'B', 'C', 'C'], 'Value': [1,2,3,4,5,6]})
data
```

```python
sns.lineplot(data, x='Category', y='Value')
```

```python
sns.barplot(data, x='Category', y='Value')
plt.show()
```

```python
sns.scatterplot(data, x='Category', y='Value')
```

```python
sns.histplot(data, x='Category', y='Value')
```

```python
sns.violinplot(data, x='Category', y='Value', color='green')
```

```python
data.plot(kind='bar', stacked=True, color = ['orange', 'green'])
plt.show()
```
