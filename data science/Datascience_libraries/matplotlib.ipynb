## Line Plot

```python
# Define the data for the x and y axes
x = [1, 2, 3, 4, 5]
y = [5, 3, 2, 6, 8]

# Plot y versus x
plt.plot(x, y)

# Display the plot
plt.show()
```

## Bar Chart

```python
# Define the data for the x and y axes
x = [1, 2, 3, 4, 5]
y = [5, 3, 2, 6, 8]

# Print the values of x and y
print("x =", x, "y =", y)

# Add title and axis labels
plt.title("X vs Y Bar Chart")  # Set the title of the plot
plt.xlabel('X-axis')           # Set the label for the x-axis
plt.ylabel('Y-axis')           # Set the label for the y-axis

# Create a bar chart
# - `x` are the x-coordinates of the bars
# - `y` are the heights of the bars
# - `color="black"` sets the color of the bars to black
# - `edgecolor="red"` sets the color of the edges of the bars to red
plt.bar(x, y, color="black", edgecolor="red")

# Display the plot
plt.show()
```

## histogram

```python
# Generate an array of 500 random numbers from a standard normal distribution
data = np.random.randn(500)

# Create a histogram of the data
# - `data` is the array of values to be binned
# - `bins=20` specifies that the data should be divided into 20 bins
# - `edgecolor="black"` sets the color of the edges of the bins to black
plt.hist(data, bins=20, edgecolor="black")
```

```python
# Create a histogram of the same data using Seaborn
# - `data` is the array of values to be binned
# - `bins=20` specifies that the data should be divided into 20 bins
# - `color="orange"` sets the color of the bins to orange
# - `edgecolor="black"` sets the color of the edges of the bins to black
# - `kde=True` overlays a Kernel Density Estimate (KDE) line on the histogram
sns.histplot(data, bins=20, color="orange", edgecolor="black", kde=True)
plt.show()
```

## Subplot
Subplots in Matplotlib allow you to create multiple plots within a single figure. This is useful for comparing different datasets or visualizations side by side.

plt.subplot(nrows, ncols, index)

* nrows: Number of rows of subplots
* ncols: Number of columns of subplots
* index: Position of the subplot in the grid, counted from left to right, top to bottom (starting from 1)

```python
# Create a 1x2 grid of subplots and select the 1st position
plt.subplot(1, 2, 1)  # 1st position

# Generate an array of 500 random numbers from a standard normal distribution
data = np.random.randn(500)

# Create a histogram of the data
# - `data` is the array of values to be binned
# - `bins=20` specifies that the data should be divided into 20 bins
# - `edgecolor="black"` sets the color of the edges of the bins to black
plt.hist(data, bins=20, edgecolor="black")

# Create a 1x2 grid of subplots and select the 2nd position
plt.subplot(1, 2, 2)  # 2nd position

# Create a histogram of the same data using Seaborn
# - `data` is the array of values to be binned
# - `bins=20` specifies that the data should be divided into 20 bins
# - `color="orange"` sets the color of the bins to orange
# - `edgecolor="black"` sets the color of the edges of the bins to black
# - `kde=True` overlays a Kernel Density Estimate (KDE) line on the histogram
sns.histplot(data, bins=20, color="orange", edgecolor="black", kde=True)

# Display the plot
plt.show()
```

```python
# Create a 3x3 grid of subplots and select the 1st position
plt.subplot(3, 3, 1)  # 1st position

# Generate an array of 500 random numbers from a standard normal distribution
data = np.random.randn(500)

# Create a histogram of the data
# - `data` is the array of values to be binned
# - `bins=20` specifies that the data should be divided into 20 bins
# - `edgecolor="black"` sets the color of the edges of the bins to black
plt.hist(data, bins=20, edgecolor="black")

# Create a 3x3 grid of subplots and select the 9th position
plt.subplot(3, 3, 9)  # 9th position

# Create a histogram of the same data using Seaborn
# - `data` is the array of values to be binned
# - `bins=20` specifies that the data should be divided into 20 bins
# - `color="orange"` sets the color of the bins to orange
# - `edgecolor="black"` sets the color of the edges of the bins to black
# - `kde=True` overlays a Kernel Density Estimate (KDE) line on the histogram
sns.histplot(data, bins=20, color="orange", edgecolor="black", kde=True)

# Display the plot
plt.show()
```

## Legend

```python
# Generate values for the x-axis
x = np.linspace(0, 10, 100)

# Compute the sine of each value in x
y = np.sin(x)

# Compute the cosine of each value in x
z = np.cos(x)

# Plot the sine function
plt.plot(x, y, label='sin of x')

# Plot the cosine function
plt.plot(x, z, label='cos of x')

# Set the x-axis label
plt.xlabel('X - Axis Label')

# Set the y-axis label
plt.ylabel('Y - Axis Label')

# Set the title of the plot
plt.title('Sin & Cos Graph')

# Add grid lines to the plot
plt.grid()

# Add a legend to the plot, positioning it in the upper right corner
plt.legend(loc='upper right')

# Display the plot
plt.show()
```

```python
# Generate values for the x-axis
x = np.linspace(0, 10, 100)

# Compute the sine of each value in x
y = np.sin(x)

# Compute the cosine of each value in x
z = np.cos(x)

# Plot the sine function
plt.plot(x, y, label='sin')

# Plot the cosine function
plt.plot(x, z, label='cos')

# Set the label for the x-axis
plt.xlabel('X axis label')

# Set the label for the y-axis
plt.ylabel('Y axis label')

# Set the title of the plot
plt.title('Plot Graph')

# Set the limit for the x-axis
plt.xlim(0, 10)

# Set the limit for the y-axis
plt.ylim(-1, 1)

# Set custom ticks for the x-axis
plt.xticks(np.arange(0, 11, 2))

# Set custom ticks for the y-axis
plt.yticks([-1, 0, 1], ['Min', 'mid', 'Max'])

# Adjust layout to prevent overlap
plt.tight_layout()

# Add grid lines to the plot
plt.grid()

# Add a legend to the plot, placing it in the upper-right corner
plt.legend(loc='upper right')

# Display the plot
plt.show()
```
