## Simple Linear Regression

•  Definition: Simple linear regression is a statistical method used to model the relationship between two variables: a dependent variable (often denoted as y) and an independent variable (often denoted as x). It assumes that there is a linear relationship between x and y, which can be represented by a straight line (regression line).

•  Purpose: The goal of simple linear regression is to find the best-fitting line (regression line) that minimizes the sum of squared differences between the observed values of y and the values predicted by the line.

•  Equation: The equation for a simple linear regression model is typically represented as: y=mx+c where m is the slope of the regression line (indicating the rate of change of y with respect to x) and c is the intercept (the value of y when x=0).

•  Usage: Simple linear regression is used to understand and predict the behavior of the dependent variable y based on changes in the independent variable x. It is widely used in various fields, including economics, social sciences, engineering, and more.

In this example, we'll demonstrate how to perform simple linear regression using scikit-learn. We'll create a dataset with two variables: study hours and the score of the student. Then, we'll use linear regression to find the best-fit line for the data. Finally, we'll visualize the data points and the regression line to understand the intuition behind linear regression.

```python
# Importing the numpy library for numerical operations
import numpy as np

# Importing the matplotlib.pyplot module for plotting graphs
import matplotlib.pyplot as plt

# Importing the LinearRegression class from the scikit-learn library for linear regression analysis
from sklearn.linear_model import LinearRegression
```

```python
# Define the dataset

# Create a numpy array for study hours and reshape it to a 2D array with one column
# This is the independent variable (X) representing the number of hours studied
study_hours = np.array([3, 5, 2, 7, 4, 6, 1, 3, 5, 4]).reshape(-1, 1)

# Create a numpy array for scores
# This is the dependent variable (y) representing the scores achieved
scores = np.array([75, 82, 69, 88, 78, 85, 62, 72, 80, 76])
```

```python
# Plot a scatter plot of study hours vs scores
plt.scatter(study_hours, scores)

# Set the label for the x-axis
plt.xlabel("Study Hours")

# Set the label for the y-axis
plt.ylabel("Scores")

# Set the title of the plot
plt.title("Scores vs Study Hours")

# Display the plot
plt.show()
```

```python
# Initialize and fit the linear regression model
# Create an instance of the LinearRegression class
model = LinearRegression()

# Fit the linear regression model using the study hours (X) and scores (y)
model.fit(study_hours, scores)
```

```python
# Retrieve the slope (coefficient) of the linear regression line
slope = model.coef_[0]

# Retrieve the intercept of the linear regression line
intercept = model.intercept_

# Display the slope and intercept
slope, intercept
```

```python
# m * x + c
slope * 2 + intercept
```

```python
# Retrieve the study hours value at index 2 (third element in the array)
actual_study_hours = study_hours[2]

# Predict the score for 2 hours of study using the fitted linear regression model
predicted_score = model.predict(np.array([2]).reshape(1, 1))

# Display both the actual study hours value at index 2 and the predicted score for 2 hours of study
actual_study_hours, predicted_score
```

```python
# Predict scores for each study hour using the fitted linear regression model
predicted_scores = model.predict(study_hours)

# Display both the original study hours array and the corresponding predicted scores
study_hours, predicted_scores
```

```python
# Scatter plot of the actual data points
plt.scatter(study_hours, scores, color="green", label="Data Points")

# Plot the linear regression line
plt.plot(study_hours, model.predict(study_hours), color="red", label="Regression Line")

# Optional: Plot a line with a slope of 1 for reference
# plt.plot(np.array([0, 10]).reshape(-1,1), np.array([0, 10]).reshape(-1,1), color="black", label="Slope Line")

# Set title and labels
plt.title('Linear Regression Example')
plt.xlabel('Study Hours')
plt.ylabel('Score')

# Display legend in the lower right corner
plt.legend(loc="lower right")

# Show the plot
plt.show()
```
